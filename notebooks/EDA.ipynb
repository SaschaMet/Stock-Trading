# Exploring yfinance API

```python
msft = yf.Ticker("MSFT")
```

```python
data = yf.download(
    ['MSFT'],
    start='2022-01-01',
    end='2022-03-01',
    interval='1d',
    actions=False,
    prepost=True,
    auto_adjust=True,
    back_adjust=True,
    index_as_date=True,
    threads=True,
)
print("Shape:", data.shape)
data.head()
```

```python
data['Close'].plot(figsize=(10, 6))
```
