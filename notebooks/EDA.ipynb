```python
import pandas as pd
import yfinance as yf
import json
import pandas as pd
import glob

#Visualisation imports
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

# Exploring yfinance API

```python
msft = yf.Ticker("MSFT")
```

```python
print(json.dumps(msft.info, indent=4, sort_keys=True))
```

```python
msft.financials
```

```python
data = yf.download(
    ['MSFT'],
    start='2022-01-01',
    end='2022-03-01',
    interval='1d',
    actions=False,
    prepost=True,
    auto_adjust=True,
    back_adjust=True,
    index_as_date=True,
    threads=True,
)
print("Shape:", data.shape)
data.head()
```

```python
data['Close'].plot(figsize=(10, 6))
```

```python
data = yf.download(
    ['^GDAXI'],
    start='2022-01-01',
    end='2022-03-01',
    interval='1d',
    actions=True,
    prepost=True,
    auto_adjust=True,
    back_adjust=True
)
print("Shape:", data.shape)
data.head()
```

```python
data.tail()
```

```python
data['Close'].plot(figsize=(10, 6))
```
